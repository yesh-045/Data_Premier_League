```python
# Merge all tables in a structured manner
fully_integrated_data = (
    pd.merge(races_df, results_df, how='inner', on=['raceId'])  # Merge race results with races
    .pipe(lambda df: pd.merge(df, constructors_df, how='left', on=['constructorId']))  # Merge constructors data
    .pipe(lambda df: pd.merge(df, constructors_standings_df, how='left', on=['raceId', 'constructorId']))  # Merge constructor standings
    .pipe(lambda df: pd.merge(df, constructors_result_df, how='left', on=['raceId', 'constructorId']))  # Merge constructor results
    .pipe(lambda df: pd.merge(df, drivers_df, how='left', on=['driverId']))  # Merge drivers data
    .pipe(lambda df: pd.merge(df, driver_standings_df, how='left', on=['raceId', 'driverId']))  # Merge driver standings
    .pipe(lambda df: pd.merge(df, circuits_df, how='left', on=['circuitId']))  # Merge circuits data
    .pipe(lambda df: pd.merge(df, status_df, how='left', on=['statusId']))  # Merge status data
    .drop(columns=[ 'statusId',  'circuitRef'])  # Drop unnecessary columns
)

# Display the final integrated dataset
fully_integrated_data.head()
```

```python
#Changing data types
data['driverDateOfBirth']=pd.to_datetime(data['driverDateOfBirth'],errors='coerce')
data['raceDate']=pd.to_datetime(data['raceDate'],errors='coerce')  
data['driverFastestLapSpeed'] = pd.to_numeric(data['driverFastestLapSpeed'],errors='coerce')
data['driverFinalGridPos'] = pd.to_numeric(data['driverFinalGridPos'],errors='coerce')
data['driverFatestLapNumber'] = pd.to_numeric(data['driverFatestLapNumber'],errors='coerce')
data.info()
```

```python
from datetime import datetime

#Adding new features
data['driverName'] = data['driverForename'] + " " + data['driverSurname']
data = data.drop(columns=['driverForename', 'driverSurname'])
data.head(5)

today = datetime.today()
data['driverAge'] = data['driverDateOfBirth'].apply(lambda dob: (today - dob).days / 365).round()
```

```python
# 1. Handle FastestLapSpeed with better null handling
data['driverFastestLapSpeed'] = (data.groupby('driverId')['driverFastestLapSpeed']
                                .transform(lambda x: x.fillna(x.median() if len(x.dropna()) > 0 else data['driverFastestLapSpeed'].median())))

# 2. Handle race position columns
position_columns = [
    'driverFinalRank',
    'driverFinalGridPos'
]

for col in position_columns:
    max_pos = data[col].max()
    data[col] = data[col].fillna(max_pos + 1)

# 3. Handle lap numbers
data['driverFatestLapNumber'] = (data.groupby('raceId')['driverFatestLapNumber']
                                .transform(lambda x: x.fillna(x.median() if len(x.dropna()) > 0 else 0)))

# 4. Championship standings with updated methods
championship_columns = {
    'constructor': [
        'constructorChampionshipStandingWins',
        'constructorChampionshipStandingPoints',
        'constructorChampionshipStandingPosition'
    ],
    'driver': [
        'driverChampionshipStandingWins',
        'driverChampionshipStandingPoints',
        'driverChampionshipStandingPosition'
    ]
}

# Updated forward fill method
for category, cols in championship_columns.items():
    for col in cols:
        data[col] = (data.sort_values('raceId')
                    .groupby(f'{category}Id')[col]
                    .transform(lambda x: x.ffill().fillna(0)))

```

```python
# Function to convert "M:SS.mmm" to seconds
def convert_to_seconds(time_str):
    if pd.isna(time_str): 
        return np.nan  
    minutes, rest = time_str.split(':')
    seconds = float(rest) 
    return int(minutes) * 60 + seconds 


def convert_to_lap_time(seconds):
    if pd.isna(seconds):
        return np.nan
    minutes = int(seconds // 60)
    sec_millis = seconds % 60
    return f"{minutes}:{sec_millis:.3f}" 

# Convert Fastest Lap Time to seconds
data['driverFastestLapTime_seconds'] = data['driverFastestLapTime'].apply(convert_to_seconds)

# Fill missing values with the mean lap time per driver
data['driverFastestLapTime_seconds'] = data.groupby('driverId')['driverFastestLapTime_seconds'].transform(lambda x: x.fillna(x.mean()))
data['driverFastestLapTime_seconds'] = data['driverFastestLapTime_seconds'].fillna(data['driverFastestLapTime_seconds'].median())
 

# Convert back to "M:SS.mmm" format
data['driverFastestLapTime'] = data['driverFastestLapTime_seconds'].apply(convert_to_lap_time)

data.drop(columns=['driverFastestLapTime_seconds'], inplace=True)

data['driverFastestLapTime'].isna().sum()
```

```python
def convert_race_time_to_seconds(time_str):
    """Convert race time to seconds, handling both full times and differences"""
    if pd.isna(time_str):
        return np.nan
    
    try:
        if isinstance(time_str, str):
            if time_str.startswith('+'):  # Time difference format
                return float(time_str.strip('+'))
            else:  # Full time format HH:MM:SS.mmm
                parts = time_str.split(':')
                if len(parts) == 3:
                    h, m, s = parts
                    return float(h) * 3600 + float(m) * 60 + float(s)
        return np.nan
    except:
        return np.nan

def process_race_times(df):
    """Process race times for each race, converting differences to full times"""
    def convert_times_for_race(race_group):
        # Find winner's time (first classified position)
        winner_mask = race_group['driverFinalRank'] == 1
        if not winner_mask.any():
            return race_group
            
        winner_time = convert_race_time_to_seconds(race_group.loc[winner_mask, 'time'].iloc[0])
        if pd.isna(winner_time):
            return race_group
            
        # Convert each time
        def convert_time(row):
            if pd.isna(row['time']):
                return np.nan
                
            time_seconds = convert_race_time_to_seconds(row['time'])
            if pd.isna(time_seconds):
                return np.nan
                
            # If it's a difference (less than 1000 seconds), add to winner's time
            if time_seconds < 1000:  # Assume differences are less than 1000 seconds
                return winner_time + time_seconds
            return time_seconds
            
        race_group['race_time_seconds'] = race_group.apply(convert_time, axis=1)
        return race_group

    # Process each race separately
    return df.groupby('raceId').apply(convert_times_for_race)

def seconds_to_race_time(seconds):
    """Convert seconds back to race time format"""
    if pd.isna(seconds):
        return np.nan
    
    hours = int(seconds // 3600)
    minutes = int((seconds % 3600) // 60)
    seconds = seconds % 60
    return f"{hours:02d}:{minutes:02d}:{seconds:06.3f}"

# Process the race times
data = process_race_times(data)

# Fill missing values using hierarchical approach
data['race_time_seconds'] = (
    data.groupby(['circuitId'])['race_time_seconds']
    .transform(lambda x: x.fillna(x.median()))
)

# Convert back to time format
data['standardized_time'] = data['race_time_seconds'].apply(seconds_to_race_time)

data['race_time'] = data['standardized_time']
data.drop(columns=['standardized_time','time','race_time_seconds'], inplace=True)
#
```
